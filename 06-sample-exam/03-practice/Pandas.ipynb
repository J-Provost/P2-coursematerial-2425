```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = pd.Series([1, 3, 2, 4])
data.plot(title="Simple Line Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({
    "Year": [2020, 2021, 2022],
    "Sales": [100, 150, 120],
    "Profit": [20, 40, 35]
})
df.set_index("Year").plot(kind="bar", title="Sales vs Profit")
plt.ylabel("Amount")
plt.show()
```

### <u> Counting birds part 3 </u>

### Bar chart of total
Display the data  using a bar chart (total) and a stacked bar chart (pigeons, sparrows, others). Ensure the total is displayed on the right side and the stacked bar on the left.
Make it informative (and pretty).

```python
# Your code
plt.figure(figsize=(10, 6))
bar_width = 0.2
index = range(len(df['year']))
plt.bar(index, df['Pigeons'], bar_width, color='blue', label='Pigeons')
plt.bar(index, df['Sparrows'], bar_width, bottom=df['Pigeons'], color='red', label='Sparrows')
plt.bar(index, df['others'], bar_width, bottom=df['Sparrows']+df['Pigeons'],color='green', label='Others')
plt.bar([i + bar_width for i in index], df['Pigeons']+df['Sparrows']+df['others'], bar_width, color='purple', label='Total')
plt.title('Birds Count Over Years')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.xticks([i + bar_width/2 for i in index], df['year'])
plt.legend()
plt.grid(axis='y') 
plt.show()
```

### <u> Counting birds part 4 </u>

### Grouped bar chart

Display the birds data using a grouped bar chart. Place the count of the pigeons and sparrows on the left of the x-tick, others and total on the right. Ensure the bars do not overlap.

```python
# Your code
plt.figure(figsize=(10, 6))
bar_width = 0.2
index = range(len(df['year']))
pigeon=plt.bar(index, df['Pigeons'], bar_width, color='blue', label='Pigeons')
sparrows=plt.bar([i + bar_width for i in index], df['Sparrows'], bar_width, color='red', label='Sparrows')
others=plt.bar([i + 2*bar_width for i in index], df['others'], bar_width, color='green', label='Others')
total=plt.bar([i + 3*bar_width for i in index], df['Pigeons']+df['Sparrows']+df['others'], bar_width, color='yellow', label='Total')
plt.bar_label(pigeon)
plt.bar_label(sparrows)
plt.bar_label(others)
plt.bar_label(total)
plt.title('Birds Count Over Years')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.xticks([i + bar_width*1.5 for i in index], df['year'])
plt.legend()
plt.grid(axis='y') 
plt.show()
```
